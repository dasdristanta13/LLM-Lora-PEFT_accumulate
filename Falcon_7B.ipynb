```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch

model = "tiiuae/falcon-7b-instruct"


```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
# init_empty_weights=False
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    
)
```

```python
sequences = pipeline(
   "Girafatron is a chatbot who knows everything about Named Entity Recognition. \nDaniel: Hello, Girafatron!Find specialty in the following sentence.\nGirafatron:Please give me the sentence. \nDaniel: 'Who is a Cardiologist?'.\nGirafatron: ",
    max_length=500,
    do_sample=True,
    top_k=20,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
)
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

# More sophisticated

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import transformers
import torch

model = "tiiuae/falcon-7b-instruct"
```

```python
from langchain.llms import HuggingFacePipeline
from langchain import PromptTemplate, LLMChain
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)


## Unable to download base_model so going to download the pipeline at once
# base_model = AutoModelForCausalLM.from_pretrained(
#     model,
#     load_in_8bit=True,
#     device_map="auto",
#     trust_remote_code=True,
#     torch_dtype=torch.bfloat16
# )
```

```python

# init_empty_weights=False
pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    # load_in_8bit=True,
# Extra Hyperparameters
    max_length=256,
    temperature = 0.6,
    top_p=0.95,
    repetition_penalty=1.2
    
)
```

```python
local_llm = HuggingFacePipeline(pipeline=pipeline)
```

```python
template = """Below is an instruction that describes a task. Write a response that appropriately complete the request.

### Instruction:
{instruction}

Answer:"""

prompt = PromptTemplate(template=template, input_variables=["instruction"])
```

```python
llm_chain=LLMChain(prompt=prompt,
                   llm=local_llm
                   )
```

```python
question = "What is the capital of England?"

print(llm_chain.run(question))
```

## Setting up chat with memory

```python
from langchain.chains import ConversationChain
from langchain.chains.conversation.memory import ConversationBufferWindowMemory

window_memory = ConversationBufferWindowMemory(k=5)
```

```python
conversation= ConversationChain(
    llm=local_llm,
    verbose=True,
    memory=window_memory
)
```

```python
conversation.prompt.template="""The following is a friendly conversation between a human and an AI, whose name is FalconChat. FalconChat is talkative and provides lots of specific details from its context. If FalconChat does not know the answer to a question, it truthfully says it does not know.

Current conversation:
{history}
Human: {input}
AI:"""
```
